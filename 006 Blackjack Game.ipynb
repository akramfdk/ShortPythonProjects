<h1>LET'S PLAY BLACKJACK !!!</h1>

```python
import random
from IPython.display import clear_output


# Initialize the deck, player_cards and computer_cards as global variables
deck = []
player_cards = []
computer_cards = []


def get_fresh_deck():
    """function is called to create a new deck for each new game"""
    return ["Ace", "Two", "Three", "Four", "Five", "Six", "Seven", "Eight", "Nine", "Ten", "Jack", "Queen", "King"]*4


def get_card_val(card):
    """returns the value of card except for Ace"""
    card_val = {
        "Two": 2, 
        "Three": 3, 
        "Four": 4, 
        "Five": 5, 
        "Six": 6, 
        "Seven": 7, 
        "Eight": 8, 
        "Nine": 9, 
        "Ten": 10, 
        "Jack": 10, 
        "Queen": 10, 
        "King": 10
    }
    
    return card_val[card]


def draw_card():
    """randomly picks a card from the deck and returns it"""
    card = random.choice(deck)
    deck.remove(card)
    
    return card


def start_game():
    """Draws two cards for both player and computer at the start"""
    
    deck.clear()
    deck.extend(get_fresh_deck())
    player_cards.clear()
    computer_cards.clear()
    
    for _ in range(2):
        player_cards.append(draw_card())
    
    for _ in range(2):
        computer_cards.append(draw_card())

        
def count_hand(cards):
    """Counts the value of the cards in a hand"""
    
    hand = 0
    ace_count = 0
    
    for card in cards:
        if card == "Ace":
            ace_count += 1
        else:
            hand += get_card_val(card)
    
    if ace_count > 0:
        if 21 - hand < 11:
            hand += ace_count
        else:
            if hand + 11 + (ace_count - 1) <= 21:
                hand = hand + 11 + (ace_count - 1)
            else:
                hand += ace_count
                
    return hand        


def player_turn():
    while count_hand(player_cards) < 21:
        player_choice = input("Type 'y' to get another card, type 'n' to pass: ")
        if player_choice == "y":  # player draws a card
            player_cards.append(draw_card())
            print(f"\tYour cards: {player_cards}, current score: {count_hand(player_cards)}")
            print(f"\tComputer's first card: {computer_cards[0]}")
        else:
            break
                
def computer_turn():
    """after player stays, computer plays till it has 17 or more score"""
    while count_hand(computer_cards) <= 16:
        computer_cards.append(draw_card())

        
def get_the_result():
    """computes the result from the final score of computer and player"""
        
    player_score = count_hand(player_cards)
    computer_score = count_hand(computer_cards)
    
    print("-"*25,"Final Score","-"*25)
    print(f"\tYour final hand: {player_cards}, final score: {count_hand(player_cards)}")
    print(f"\tComputer's final hand: {computer_cards}, final score: {count_hand(computer_cards)}")
    
    if player_score > 21:
        print("You went over. You lose \U0001F612")
    else:
        if computer_score > 21:
            print("Opponent went over. You win \U0001F601")
        elif computer_score > player_score:
            print("You lose \U0001F612")
        elif computer_score == player_score:
            print("Draw \U0001F643")
        else:
            print("You win \U0001F601")
    
    print("*"*100)
    
def play_game():
    """This function controls the play of the game"""
    play_game = "y"

    while play_game == "y":
        start_game()
        print(f"\tYour cards: {player_cards}, current score: {count_hand(player_cards)}")
        print(f"\tComputer's first card: {computer_cards[0]}")

        player_turn()
        computer_turn()
        get_the_result()
        
        play_game = input("Do you want to play a game of Blackjack? Type 'y' or 'n': ")

    print("The Game Ends..GoodBye !!!")
    
play_game()
```
