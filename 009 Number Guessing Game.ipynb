# Number Guessing Game

```python
import random

logo = '''
________                             __________.__            
 /  _____/ __ __  ____   ______ ______ \__    ___|  |__   ____  
/   \  ___|  |  _/ __ \ /  ___//  ___/   |    |  |  |  \_/ __ \ 
\    \_\  |  |  \  ___/ \___ \ \___ \    |    |  |   Y  \  ___/ 
 _______  |____/ \___  ___._  /____  >   |___._._._._|  /\___  >
 \      \/ __ __  _____\_ |__/  ___________  | | | |  \/     \/ 
 /   |   \|  |  \/     \| __ \_/ __ \_  __ \ | | | |            
/    |    |  |  |  Y Y  | \_\ \  ___/|  | \/  \|\|\|            
\____|__  |____/|__|_|  |___  /\___  |__|     ______            
        \/            \/    \/     \/         \/\/\/
'''

def welcome():
    print(logo)
    print("Welcome to the Number Guessing Game!")
    

def get_attempts():
    """asks the user for difficulty level - easy or hard
    for easy level - 10 attempts
    for hard level - 5 attempts"""
    
    while True:
        difficulty = input("Choose a difficulty. Type 'easy' or 'hard': ").lower()
        if not(difficulty == 'easy' or difficulty == 'hard'):
            print("Invalid Input. Try Again !!!")
        else:
            break
            
    attempts = 0
    if difficulty == "easy":
        attempts = 10
    else:
        attempts = 5
        
    return attempts


def play_game():
    welcome()
    
    secret_num = random.randint(1, 100)
    print("I'm thinking of a number between 1 and 100.")
    
    attempts = get_attempts()
    
    while attempts > 0:
        print(f"You have {attempts} attempts remaining to guess the number.")
        guess = int(input("Make a guess: "))
        
        if guess > secret_num:
            print("Too high.\nGuess again.")
        elif guess < secret_num:
            print("Too low.\nGuess again.")
        else:
            print(f"You got it! The answer was {secret_num}.")
            break
        
        attempts -= 1
            
    if attempts == 0:
        print("You've run out of guesses, you lose.")
        
play_game()
```
