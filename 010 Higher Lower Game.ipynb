```python
def ask_user():
    while True:
        user_choice = input("Who has more followers? Type 'A' or 'B': ").upper()
        if user_choice == "A" or user_choice == "B":
            break
        else:
            print("Invalid Input. Try Again!!!")
    
    return user_choice


def check_user_choice(samples, user_choice):
    if samples[0]["follower_count"] > samples[1]["follower_count"] and user_choice == "A":
        return True
    elif samples[0]["follower_count"] < samples[1]["follower_count"] and user_choice == "B":
        return True
    else:
        return False

def play_game():
    # print the logo
    print(logo)

    score = 0
    while True:
        # pick two random samples from data
        two_samples = random.sample(data, 2)

        # print the samples for the user to see
        print(f"Compare A: {two_samples[0]['name']}, a {two_samples[0]['description']}, from {two_samples[0]['country']}.")
        print(vs)
        print(f"Against B: {two_samples[1]['name']}, a {two_samples[1]['description']}, from {two_samples[1]['country']}.")

        # ask the user to choose the sample with more views
        user_choice = ask_user()

        # check if the user is correct and if so increase the score, print it and repeat
        if check_user_choice(two_samples, user_choice):
            score += 1
            print(f"You're right! Current score: {score}.")
        else:
            break

    # if incorrect, exit the game and print final score
    print(f"Sorry, that's wrong. Final score: {score}")

play_game()
```
